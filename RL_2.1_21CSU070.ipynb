CORRELATION

```python
x=np.random.randint(0,50,1000)
y=x+np.random.normal(0,10,1000)
```

# if matrix is towards 1 then +ve correlation and visa versa

```python
np.corrcoef(x,y)
```

```python
plt.scatter(x,y)
plt.show()
```

# Here 100 is been subtracted

```python
y=100-(x+np.random.normal(0,10,1000))
y
```

```python
np.corrcoef(x,y)
```

```python
plt.scatter(x,y)
plt.show()
```

CREATE A DATAFRAME

```python
df=pd.DataFrame({'a':np.random.randint(0,50,1000)})
df['b']=df['a']+np.random.normal(0,10,1000)
df['c']=100-df['a']+np.random.normal(0,5,1000)
df['d']=np.random.randint(0,50,1000)

df.corr()
```

```python
pd.plotting.scatter_matrix(df,figsize=(100,100))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)),df.columns)
plt.yticks(range(len(df.columns)),df.columns)
plt.colorbar()
plt.show()
```
